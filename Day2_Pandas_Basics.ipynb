```python
import pandas as pd
import seaborn as sns

df=sns.load_dataset("titanic")

df.head()
```

```python
print("Shape:",df.shape)


print("Columns:",df.columns)

df.info()


df.describe(include="all")
```

```python
df['age'].head()

df[['age', 'sex', 'fare' ]].head()


df[df['age']>60]


df[(df['age']>60)&(df['sex']=="male")&(df['survived']==1)]
```

```python
df.sort_values("fare", ascending=False).head()

df.groupby("sex")['age'].mean()

df.groupby("class")['survived'].mean()
```

```python
pivot=df.pivot_table(values="survived", index="class", columns="sex", aggfunc="mean")
print(pivot)
```
