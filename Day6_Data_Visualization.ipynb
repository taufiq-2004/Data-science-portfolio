```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=sns.load_dataset('titanic')

plt.plot([1,2,3,4], [2,4,6,8], marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


plt.bar(['A', 'B', 'C'], [10,15,20])
plt.title("Bar Plot")
plt.show()
```

```python
plt.hist(df['age'].dropna(), bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.histplot(df['age'],bins=20, kde=True)
plt.title("Age Distribtion with KDE")
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title("Count of Male vs Female")
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=df)
plt.title("Age Distribution by Class")
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=df)
plt.title("Violin plot: Age vs Sex")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='sex', data=df)
plt.title("Age vs Fare (by sex)")
plt.show()
```

```python
plt.hist(df['fare'].dropna(), bins=30, edgecolor="black")
plt.title("Passenger fare distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival Count By Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
sns.boxplot(x="class", y="fare" ,data=df)
plt.title("Fare Distribution by Class")
plt.show()
```

```python
sns.violinplot(x="survived", y="age", data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.scatterplot(x="age", y="fare", hue="class", data=df, alpha=0.5)
plt.title("Age vs Fare(by class)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
