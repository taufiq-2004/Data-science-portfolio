```python
import pandas as pd
import seaborn as sns

df=sns.load_dataset("titanic")
df.head()


```

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("class")[["fare", "survived"]].mean()
```

```python
df.groupby(['class', 'sex'])['survived'].mean()
```

```python
df.groupby('class')['survived'].count()
```

```python
data1=pd.DataFrame({
    "id": [1,2,3],
    "name": ["Alice", "Bob", "Charlie"]
})

data2=pd.DataFrame({
    "id": [1,2,3],
    "age": [25,30,35]

})

merged=pd.merge(data1, data2, on="id")
print(merged)
```

```python
concat_df=pd.concat([data1, data2], axis=0)
print(concat_df)
```

```python
pd.pivot_table(df, values="fare", index="class", columns="sex", aggfunc="mean")

pd.crosstab(df['class'], df['sex'])
```

```python
df['age_group']=df['age'].apply(lambda x: "Child" if x<18 else "Adult")
df[['age', 'age_group']].head()

def tip_status(fare):
  if fare>100:
    return "High fare"
  elif fare>50:
    return "Medium Fare"
  else:
    return "Low Fare"

df['fare_level']=df['fare'].apply(tip_status)
df[['fare', 'fare_level']].head()
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('class')['fare'].mean()
```

```python
df.groupby(['class', 'sex'])['survived'].sum()
```

```python
pd.pivot_table(df, values='age', index='class', columns='survived', aggfunc='mean')
```

```python
def fare_category(fare):
  if fare < 20:
    return "Low Fare"
  elif fare <= 100:
    return "Medium Fare"
  else:
    return "High Fare"

df['fare_category']=df['fare'].apply(fare_category)
df[['fare', 'fare_category']].head()
```
