```python
import numpy as np

arr=np.array([1,2,3,4,5])

print("Array:", arr)


arr2=np.arange(0, 10, 2)
print("Arange", arr2)

rand_arr=np.random.rand(3,3)
print("Random 3x3 Matirx:\n",rand_arr)
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print("a+b", a+b)
print("a*b", a*b)
print("dot:", np.dot(a,b))
print("mean:", a.mean(), "std:", a.std())
```

```python
m=np.arange(1,13).reshape(3,4)
print("Matrix:\n", m)


print("First Row", m[0])
print("First Columns", m[:, 0])

print("Sub-Matrix", m[0:2, 1:3])
```

```python
A=np.array([[1,2], [3,4]])
B=np.array([[5,6], [7,8]])


print("Matrix Multiplication", A @ B)
print("Transpose", A.T)
print("Determinant", np.linalg.det(A))
print("Inverse", np.linalg.inv(A))
```

```python
import matplotlib.pyplot as plt

x=np.linspace(0, 10, 100)
y=np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.show()



p=np.random.rand(50)
q=np.random.rand(50)

plt.scatter(p, q)
plt.title("Scatter Plot")
plt.show()

data=np.random.rand(1000)
plt.hist(data, bins=10, edgecolor="black")
plt.title("Histogram of  Random Data")
plt.show()
```
